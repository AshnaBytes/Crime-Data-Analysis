#                                    ***Exploratory Data Analysis (EDA) of Crime Data***


## **1. Introduction**


### 1.1 Overview of Dataset:

This dataset contains crime records from 2020 to the present, including crime type, location, date, time, and victim demographics. This analysis aims to understand crime trends, identify high-crime areas, and explore the nature of reported crimes.

### 1.2 Purpose of EDA:

The goal is to clean the data, visualize patterns, and derive insights that could be useful for law enforcement, policy-making, and crime prevention strategies.

## **2. Data Preparation**

### 2.1 Importing Python Libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2.2 Reading Dataset:

```python
crime= pd.read_csv("Crime_Data_from_2020_to_Present.csv")
crime.head()
```

```python
crime.tail()
```

```python
crime.shape
```

## **3. Data Cleaning**

```python
crime.info()
```

```python
crime.isna().sum()
```

```python
missing_percentage = (crime.isna().sum() / len(crime) * 100).round(2)
missing_percentage = missing_percentage[missing_percentage > 0]
print(missing_percentage)
```

* ['Crm Cd 4', 'Crm Cd 3', 'Crm Cd 2', 'Cross Street', 'Weapon Used Cd', 'Weapon Desc' ] - **Must be removed**


### 3.1 Removing Unnecessary Columns



```python
crime=crime.drop(["DR_NO","Crm Cd 2","Crm Cd 3","Crm Cd 4","Weapon Used Cd","Weapon Desc","Mocodes"],axis=1)
crime.head()
```

### 3.2 Handling missing value

```python
for cols in crime.select_dtypes(include='object').columns:
    crime[cols]=crime[cols].fillna(crime[cols].mode()[0])

crime.isna().sum()
```

```python
for cols in crime.select_dtypes(exclude='object').columns:
    crime[cols]=crime[cols].fillna(crime[cols].mean())
    
crime.isna().sum()
```

### 3.3 Data Type Conversions

*Convert object-type columns to datetime format where applicable*

```python
crime["Date Rptd"]=pd.to_datetime(crime["Date Rptd"],format='%Y/%m/%d %I:%M:%S %p')
crime["DATE OCC"]=pd.to_datetime(crime["DATE OCC"],format='%Y/%m/%d %I:%M:%S %p')

crime["Year"] = crime["DATE OCC"].dt.year
crime["Month"] = crime["DATE OCC"].dt.month
crime["WeekDay"] = crime["DATE OCC"].dt.weekday

crime["Hours"] = crime["TIME OCC"]/100

crime.head()
```

```python
crime.info()
```

## **4. Exploratory Data Analysis (EDA)**

### 4.1 Crime Trends Over Time

##### 4.1.1 Yearly Crime Analysis

```python
Number_of_Crimes = crime.groupby("Year").size().reset_index(name="Number_of_Crimes")
Number_of_Crimes
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=Number_of_Crimes, x="Year", y="Number_of_Crimes", marker='o',color="m")
plt.title("Annual Crime Statistics", fontweight="bold")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.grid(True)

plt.show()
```

* The most frequently reported crimes occurred in 2022 (235,252 cases), while 2024 saw a decline (127,525 cases).
* Possible reasons for the decline: increased law enforcement efforts, policy changes, or underreporting.

##### 4.1.2 Monthly Crime Analysis

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=crime,x="Month",color="c")
plt.title("Monthly Crime Statistics",fontweight="bold")
plt.xlabel("Month Of The Year")
plt.ylabel("Crime Count")
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

for bar in plt.gca().patches:
    plt.gca().annotate(
        int(bar.get_height()), (bar.get_x() + bar.get_width() / 2, bar.get_height()), ha='center', va='bottom', fontsize=8,color='black')
plt.show()
```

* Crime rates peak in January, whereas December sees fewer reported incidents.
* The spike in January could be due to post-holiday crime patterns, while the increase on Fridays may relate to increased social activities.

##### 4.1.3 Crime by Day of the Week

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=crime,x="WeekDay",color="olive")
plt.title("Crime Patterns Throughout the Week",fontweight="bold")
plt.xlabel("Day Of Week")
plt.ylabel("Total Crime Inccidents")
plt.xticks(range(7), ["Mon","Tue","Wed","Thur","Fri","Sat","Sun"])

for bar in plt.gca().patches:
    plt.gca().annotate(
        int(bar.get_height()), (bar.get_x() + bar.get_width() / 2, bar.get_height()), ha='center', va='bottom', fontsize=8,color='black')
plt.show()
```

* Fridays have the highest number of reported crimes.
* Mondays record the lowest crime rates.
* This trend could be linked to social activities and weekend behavior patterns

##### 4.1.4 Crime by Day of the Week

```python
plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")
sns.histplot(data=crime,x="Hours",bins=24,color="teal",kde=True)
plt.title("Crime Occurance By The Hour Of Day",fontweight="bold")
plt.xlabel("Hour Of The Day")
plt.ylabel("Total Crime Inccidents")

plt.show()
```

* Peak crime hours: 1 AM - 7 AM.
* Least crime activity: Around 5 PM.
* Possible reasons include increased police patrolling during the day and late-night criminal activities.



### 4.2 Most Common Crimes

##### 4.2.1 Top 10 Most Reported Crimes

```python
most_common_crime= crime["Crm Cd Desc"].value_counts().nlargest(10)
most_common_crime
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(y=most_common_crime.index,x=most_common_crime.values,color="coral")
plt.title("Top 10 Most Reported Crime Categories",fontweight="bold")
plt.xlabel("Total Reported Cases")
plt.ylabel("Crime Type")
plt.grid(axis="x", linestyle="--", alpha=0.6)
plt.show()
```

* In Los Angeles, the most frequently reported crimes include vehicle theft, simple assault, and burglary.
* Additionally, the city sees a notable number of cases involving identity theft and vandalism.

### 4.3 Correlation matrix among different attributes


```python
plt.figure(figsize=(15, 6))
numeric_crime = crime.select_dtypes(include=['number'])
sns.heatmap(numeric_crime.corr(),annot=True)
plt.show()
```

1. Crime severity is highly correlated with crime type and is slightly related to the age of the victim and the type of weapon used.

### 4.5 Analyzing Crime by Location

##### 4.5.1 Top 10 effected area by crime

```python
most_affected_areas = crime["AREA NAME"].value_counts().nlargest(10)
most_affected_areas
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(y=most_affected_areas.index, x=most_affected_areas.values,color="sandybrown")
plt.title("Top 10 Areas with Highest Crime Incidents", fontweight="bold")
plt.xlabel("Total Crime Cases")
plt.ylabel("Location")
plt.grid(axis="x", linestyle="--", alpha=0.6)
plt.show()
```

##### 4.5.2 Type of crime in each area


```python
filtered_crime = crime[
    (crime["AREA NAME"].isin(most_affected_areas.index)) &
    (crime["Crm Cd Desc"].isin(most_common_crime.index))
]
crime_grouped = filtered_crime.groupby(["AREA NAME", "Crm Cd Desc"]).size().reset_index(name="Count")
crime_grouped
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(data=crime_grouped, x="AREA NAME", y="Count", hue="Crm Cd Desc", dodge=False)
plt.title("Crime Count by Area and Crime Description", fontweight="bold", fontsize=14)
plt.xlabel("Area", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(loc='upper right', title="Crime Description", fontsize=10)
plt.show()
```

Vehicle theft is the most common crime in all areas.


### 4.6 Victim Demographics

##### 4.6.1 Age-wise Crime Distribution

```python
age_intervals = [0, 18, 35, 50, 65, 100]  
age_categories = ['Under 18', '19-35', '36-50', '51-65', '66 and above']

crime['Age Category'] = pd.cut(crime['Vict Age'], bins=age_intervals, labels=age_categories, right=False)
crime['Age Category']
```

```python
crime_counts_by_age = crime['Age Category'].value_counts().sort_index()
crime_counts_by_age
```

```python
plt.figure(figsize=(10,7))
plt.pie(crime_counts_by_age.values, labels=crime_counts_by_age.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("flare", len(crime_counts_by_age)))
plt.title("Crime Incidence Across Age Categories", fontsize=14)
plt.xlabel("Age Category", fontsize=12)

plt.show()
```

Young adults (19-35) are more frequently victims of crime.

##### 4.6.2 Severity acrross age category

```python
sns.countplot(data=crime, x='Age Category',hue="Part 1-2",palette='crest')
plt.title("Crime Severity Across Age Group")
plt.ylabel("Number Of Crime")
plt.show()
```

* **Young Victims (Under 25):**
  - More likely to be victims of violent crimes, such as assault and robbery.
  - High involvement in gang-related and street crimes.
* **Middle-Aged Victims (25-50):**
   - Experience a mix of violent and non-violent crimes, such as fraud, workplace-related crimes, and domestic violence.
* **Elderly Victims (50+):**
   - More prone to financial scams, fraud, and elder abuse rather than physical violence.

##### 4.6.3 Gender-based Analysis

```python
crime_by_sex= crime[~crime["Vict Sex"].isin(["H", "-"])]
gender_counts=crime_by_sex["Vict Sex"].value_counts()
gender_counts
```

```python
plt.figure(figsize=(7, 5))
sns.barplot(x=gender_counts.index, y=gender_counts.values, color="brown")
plt.title("Crime Victimization by Gender", fontweight="bold")
plt.xlabel("Gender")
plt.ylabel("Number of Victims")
plt.xticks(range(3),["M","F","OTHER"])
for index, value in enumerate(gender_counts.values):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=12)

plt.show()
```

Male are most likely to face crime then Females


##### 4.6.4 Crime Severity by Gender

```python
severity_by_sex = crime.groupby(["Vict Sex", "Part 1-2"]).size().reset_index(name="Count")
severity_by_sex
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=severity_by_sex, x="Vict Sex", y="Count", hue="Part 1-2", palette="flare")
plt.title("Crime Severity Across Victim Sex", fontweight="bold")
plt.ylabel("Number of Crimes")
plt.xlabel("Victim Sex")
plt.xticks(range(3),["M","F","OTHER"])
plt.legend(title="Crime Severity (Part 1 & 2)")
plt.show()
```

* Males are more likely to be victims of violent crimes, including aggravated assault, homicide, and robbery.
* Females experience higher victimization in domestic-related crimes, such as domestic violence, sexual assault, and stalking.


### 4.7 Crime Type and Arrests

```python
crime["Status"].head()
```

```python
crime_arrests = crime[crime["Status"] == "AA"].groupby(["Crm Cd Desc"]).size().reset_index(name="Arrest_Count")
crime_arrests
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=crime_arrests, x="Arrest_Count", y="Crm Cd Desc", color="lightskyblue")
plt.xlabel("Number of Arrests")
plt.ylabel("Crime Type")
plt.title("Top 10 Crimes with Highest Arrests", fontweight="bold")
plt.show()
```

* Crimes such as homicide, assault with a deadly weapon, and narcotics-related offenses have higher arrest rates due to stronger investigative efforts and clear legal evidence.
* Property crimes (e.g., burglary, theft, vandalism) often have lower arrest rates due to lack of direct evidence and delayed reporting.
* Fraud and cybercrimes also show low arrest rates, likely due to the complexity of tracking digital criminals.
